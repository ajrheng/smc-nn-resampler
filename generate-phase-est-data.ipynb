```python
import phase_est
import numpy as np
import matplotlib.pyplot as plt
```

```python
t0 = 0.1
omega_star = np.random.uniform() * 2 * np.pi
print(omega_star)
```

```python
smc = phase_est.phase_est_smc(omega_star, t0)
smc.init_particles(10000) # 10000 particles
```

```python
for i in range(100):
    particle_pos, particle_wgts = smc.update()
    smc.update_t()
    plt.scatter(particle_pos, particle_wgts)
    plt.show()
```
